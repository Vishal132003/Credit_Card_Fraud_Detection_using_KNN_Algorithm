```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
df=pd.read_csv("creditcard.csv")
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
scaler=StandardScaler()
df['NormalizedAmount']=scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df=df.drop(['Amount','Time'],axis=1)
```

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=16)
```

```python
# Initialize the KNeighborsClassifier model
model=KNeighborsClassifier()
# Train the model using the scaled training data
model.fit(x_train,y_train)
```

```python
# Scale the training and testing data using StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Initialize the KNeighborsClassifier model
model=KNeighborsClassifier()
# Train the model using the scaled training data
model.fit(x_train_scaled,y_train)

# Make predictions on the scaled testing data
y_pred=model.predict(x_test_scaled)
```

```python
# Generate the confusion matrix to evaluate the model's performance
confusion_matrix(y_test,y_pred)
```

```python
# Generate a classification report for detailed evaluation metrics
classification_report(y_test,y_pred)
```

```python
# Calculate the accuracy score of the model
accuracy_score(y_test,y_pred)
```

```python
# Import the pickle library to save the trained model
import pickle
# Open a file named "best.pkl" in write binary mode and save the model
with open("best.pkl","wb") as f:
  pickle.dump(model,f)
```
